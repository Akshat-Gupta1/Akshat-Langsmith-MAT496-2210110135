# Tracing Basics

## What I Learned
In this video, I learned how to use LangSmith to trace the execution of functions using the @traceable decorator from the LangSmith Python SDK. Every function call gets logged as its own run, making it easy to debug and track execution.

## Changes in Code
I set up environment variables for tracing and decorated my functions with @traceable. I updated the model configuration and customized the prompts for my use case.

```python
import os
from langsmith import traceable
from langchain_openai import ChatOpenAI

# Set up environment
os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ["LANGCHAIN_API_KEY"] = "your-api-key"

@traceable
def simple_llm_call(user_input: str) -> str:
    llm = ChatOpenAI(model="gpt-4o-mini")
    response = llm.invoke(user_input)
    return response.content

# Test the function
result = simple_llm_call("What is LangSmith?")
print(result)
```

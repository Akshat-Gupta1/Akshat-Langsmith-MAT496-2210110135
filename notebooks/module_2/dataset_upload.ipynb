# Datasets

## What I Learned
I learned how to create and manage datasets in LangSmith using the SDK. Datasets can be added programmatically, manually, or generated by AI. I created a dataset focused on general Q&A for practice.

## Changes in Code
I uploaded my dataset examples to LangSmith through the SDK and configured the environment properly. This helped me get familiar with dataset handling beyond just the LangSmith UI.

```python
from langsmith import Client
import os

# Initialize client
client = Client()

# Create a dataset
dataset_name = "qa_examples"
dataset = client.create_dataset(
    dataset_name=dataset_name,
    description="Q&A examples for testing"
)

# Add examples to dataset
examples = [
    {
        "inputs": {"question": "What is LangSmith?"},
        "outputs": {"answer": "LangSmith is a platform for debugging, testing, and monitoring LLM applications."}
    },
    {
        "inputs": {"question": "What is tracing?"},
        "outputs": {"answer": "Tracing is the process of logging execution steps of your LLM application."}
    },
    {
        "inputs": {"question": "What are datasets used for?"},
        "outputs": {"answer": "Datasets are used for evaluating and testing LLM applications consistently."}
    }
]

for example in examples:
    client.create_example(
        inputs=example["inputs"],
        outputs=example["outputs"],
        dataset_id=dataset.id
    )

print(f"Created dataset '{dataset_name}' with {len(examples)} examples")
```
